```python
# combine column and create X and Y  
pca = PCA(n_components=2)
pca_data = pca.fit_transform(feature1_df)
pca_df = pd.DataFrame(data = pca_data
             , columns = ['PCA X', 'PCA Y'])
pca_df.shape
```

```python
pca_df.head()
```

```python
# melakukan ploatting berdasarkan X dan Y
sns.scatterplot(x='PCA X', y='PCA Y', data=pca_df)
```

```python
# Inisiasi Variable
K = 29 # based on team position types.
iterations = 150

pca_df.values.shape
```

```python
# membuat centeroid
Centroids=np.array([]).reshape(pca_df.values.shape[1],0) 
Centroids
```

```python
# memilih titik centeroid secara acak
for i in range(K):
    rand=np.random.randint(0,pca_df.values.shape[0]-1)
    Centroids=np.c_[Centroids,pca_df.values[rand]]
```

```python
output = {}
```

```python
for i in range(iterations):
    # matrix for calculate euclidian distance
    EuclidianDistance=np.array([]).reshape(pca_df.values.shape[0],0)
    for i in range(K):
        tempDist=np.sum((pca_df-Centroids[:,i])**2,axis=1)
        EuclidianDistance=np.c_[EuclidianDistance,tempDist]
    C=np.argmin(EuclidianDistance,axis=1)+1

    temp={}
    for i in range(K):
        temp[i+1]=np.array([]).reshape(2,0)

    for i in range(pca_df.values.shape[0]):
        temp[C[i]]=np.c_[temp[C[i]],pca_df.values[i]]

    for i in range(K):
        temp[i+1]=temp[i+1].T

    for i in range(K):
         Centroids[:,i]=np.mean(temp[i+1],axis=0)
    output = temp
```

```python
colors = [
    'aliceblue',
     'antiquewhite',
     'aqua',
     'aquamarine',
     'azure',
     'beige',
     'bisque',
     'black',
     'blanchedalmond',
     'blue',
     'blueviolet',
     'brown',
     'burlywood',
     'cadetblue',
     'chartreuse',
     'chocolate',
     'coral',
     'cornflowerblue',
     'cornsilk',
     'crimson',
     'cyan',
     'darkblue',
     'darkcyan',
     'darkgoldenrod',
     'darkgray',
     'darkgreen',
     'darkgrey',
     'darkkhaki',
     'darkmagenta'
]
labels=[
    'cluster1','cluster2','cluster3','cluster4','cluster5',
    'cluster6','cluster7','cluster8','cluster9','cluster10',
    'cluster11','cluster12','cluster13','cluster14','cluster15',
    'cluster16','cluster17','cluster18','cluster19','cluster20',
    'cluster21','cluster22','cluster23','cluster24','cluster25',
    'cluster26','cluster27','cluster28','cluster29']
colors
```

```python
# melakukan ploting dengan scatter beserta labeling
for k in range(K):
    plt.scatter(output[k+1][:,0],output[k+1][:,1],c=colors[k],label=labels[k])
    
plt.scatter(Centroids[0,:],Centroids[1,:],s=10,c='violet',label='Centroids')
plt.xlabel('attacker_point')
plt.ylabel('keeper_point')
plt.legend()
plt.show()
```

# Compared using library

```python
from sklearn.cluster import KMeans

# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 29, init = 'k-means++', random_state = 150)
y_kmeans = kmeans.fit_predict(pca_df)

# Visualising the clusters
for i in range(29):
    plt.scatter(pca_df.values[y_kmeans == i, 0], pca_df.values[y_kmeans == i, 1], s = 100, c = colors[i], label = labels[i])
    
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=10,c='violet',label='Centroids')

plt.title('Clusters of KMeans')
plt.xlabel('attacker_point')
plt.ylabel('keeper_point')
plt.legend()
plt.show()
```
